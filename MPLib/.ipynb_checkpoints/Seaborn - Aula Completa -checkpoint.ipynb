## Principais Imports

```python
titanic = sb.load_dataset('titanic')
titanic.head(15)
```

```python
tips = sb.load_dataset('tips')
tips.head(20)
```

```python
fly = sb.load_dataset('flights')
fly.head(10)
```

## Tipos de Graficos

### Joinplot | Grafico de Correlao entre 2 Variaveis

```python
#Vamos verificar total_bill x tips

correl = sb.jointplot(x='total_bill',y='tip',data=tips)

# Calcula a correlao de Pearson
r, p = stats.pearsonr(tips['total_bill'], tips['tip'])

# Adiciona o valor da correlao(r) e da Significancia Estatistica (p) ao grfico
correl.ax_joint.annotate(f'r = {r:.2f}\np = {p:.2g}', xy=(0.1, 0.9), xycoords='axes fraction', fontsize=12)

plt.show()
```

```python
#Verificando valor da pasagem com Idade

navio = sb.jointplot(x='fare',y='age',data=titanic)

# Calcula a correlao de Pearson
#r, p = stats.pearsonr(titanic['fare'], titanic['age'])

# Adiciona o valor da correlao(r) e da Significancia Estatistica (p) ao grfico
#navio.ax_joint.annotate(f'r = {r:.2f}\np = {p:.2g}', xy=(0.1, 0.9), xycoords='axes fraction', fontsize=12)

plt.show()
```

```python
#Tamanho da mesa e tips

correl = sb.jointplot(x='size',y='tip',data=tips)


# Calcula a correlao de Pearson
r, p = stats.pearsonr(tips['size'], tips['tip'])

# Adiciona o valor da correlao(r) e da Significancia Estatistica (p) ao grfico
correl.ax_joint.annotate(f'r = {r:.2f}\np = {p:.2g}', xy=(0.05, 0.9), xycoords='axes fraction', fontsize=12)

plt.show()
```

### lmplot | Correlao entre 2 Variaveis (Joinplot) com linha para Regresso Linear

```python
tips.head()
```

```python
sb.lmplot(x='total_bill',y='tip', data=tips, hue='time')
plt.show()
```

```python
# Criar o lmplot separado por 'time'
correl2 = sb.lmplot(x='total_bill',y='tip', data=tips, hue='time', col='time')

# Calcular e anotar a o valor da correlao(r) e da Significancia Estatistica (p) em cada subplot

for ax, horario in zip(correl2.axes.flat, tips['time'].unique()):
    subset = tips[tips['time'] == horario]
    r, p = stats.pearsonr(subset['total_bill'], subset['tip'])
    ax.annotate(f'r = {r:.3f}\np = {p:.2g}', xy=(0.05, 0.9), xycoords='axes fraction', fontsize=12)

plt.show()
```
