# Trabalhando com data e tempo no Pandas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file="C:/Users/Vini Estudos/Documents/Cursos e Estudos/CursoDataScience/Projetos_DataScience/20250512 - Pandas - UFO_Nulos/ufo.csv"
ufo = pd.read_csv(file)
ufo.head()
```

```python
# Time  armazenado como string
ufo.dtypes
```

# Podemos utilizar do mtodo str j ensinado para acessar as posies da data

```python
# Porm essa abordagem no  100% segura

ufo.Time.str.slice(-5, -3).astype(int).head(10)
```

```python
# Vamos converter a coluna Time para o formato Datetime

ufo['Time'] = pd.to_datetime(ufo.Time)
ufo.head()
```

```python
ufo.dtypes
```

# Com uma coluna do tipo Datetime, temos mtodos muito melhores

```python
# Selecionando a hora
    
ufo.Time.dt.hour.head()
```

```python
#Verificando o dia da semana da data

ufo.Time.dt.weekday_name.head()
```

```python
#Verificando o dia do ano da data

ufo.Time.dt.dayofyear.head()
```

```python
#ufo.Time.dt.month_name(locale='pt_BR.utf8')
ufo["safra"] = ufo.Time.dt.to_period("M")
ufo.head(20)
```

# Como funciona a converso

```python
#O resultado  um timestamp
#Dayfirst  para mostrar ao pandas que a data come com dd/mm/yyyy, tambem  possivel passar o formato format='%d/%m/%Y'

ts = pd.to_datetime('31/1/1999',dayfirst=True)
ts
```

```python
# Comparando uma srie Datetime com Timestamp

ufo.loc[ufo.Time >= ts, :].head(20)
```

```python
#Encontrando a data mais alta

maxima = ufo.Time.max()
maxima
```

```python
#Encontrando a menor data

minima = ufo.Time.min()
minima
```

```python
# Podemos fazer operaes matemticas com as datas

td = maxima - minima
td
```

```python
# O tipo Timedelta tambm tem atributos que podem ser acessados
# Aperte o TAB

td.
```

# Criando um grfico de aparies de UFO ao longo dos anos

```python
# Utilizando o matplotlib

%matplotlib inline
```

```python
# Criando a coluna YEAR utilizando os mtodos do tipo dt

ufo['Year'] = ufo.Time.dt.year
```

```python
#Verificando a coluna

ufo.head()
```

```python
#Contando os valores de aparies de anos

ufo.Year.value_counts().head(20)
```

```python
#Observe que os valores no vem ordenados, o que invalda esse tipo de grfico

ufo.Year.value_counts().sort_index().head(10)
```

```python
ufo.Year.value_counts().sort_index().tail(10)
```

```python
# Plotando o grfico

ufo.Year.value_counts().sort_index().plot()
plt.show()
```
